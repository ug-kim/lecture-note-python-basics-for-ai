# basic, loc, iloc selection

```python
df[["name", "street"]][:2]  # column과 index number
```

`loc` - location으로, row index 이름을 명확히 적어야한다

`iloc` - 실제 index로, list처럼 사용한다

```python
df.loc[[211829, 320563], ["name", "street"]]
# df.loc[:2, ["name", "street"]] -> 에러, row index 명확하게 적어줘야함
# df.loc[:320563, ["name", "street"]] -> 가능
```

```python
df.iloc[:10, :3]  # row, column
```

# reindex 인덱스 재설정

```python
df.index = list(range(0, 15))
# df.reset_index(inplace=True)
df.head()
```

```python
df.reset_index()  # index 하나 더 생긴다, df 자체가 변하지 않는다
```

```python
df.reset_index(drop=True)  # drop=True 하면 기존 index 사라지고 새로운 인덱스만 있다
```

```python
df.reset_index(inplace=True)  # inplace=True 하면 변경된 값 들어간다, df 자체 변한다
# df 값은 유지하려고 노력, 왠만하면 바뀌지 않는다
```
