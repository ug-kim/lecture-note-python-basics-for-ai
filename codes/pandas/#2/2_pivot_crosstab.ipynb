```python
import pandas as pd
import dateutil
```

# Pivot table Crosstab

### pivot table

```python
df_phone = pd.read_csv("phone_data.csv")
df_phone["date"] = df_phone["date"].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
# pivot table 말고 groupby 사용
df_phone.groupby(["month", "item", "network"])["duration"].sum().unstack()
```

```python
df_phone.pivot_table(
    values=["duration"],
    index=[df_phone.month, df_phone.item],  # index list 형태로
    columns=df_phone.network,
    aggfunc="sum",
    fill_value=0,
)
```

### crosstab

- 특허 두 칼럼에 교차 빈도, 비율, 덧셈 등을 구할 때 사용
- pivot table의 특수한 형태
- User-Item Rating Matrix 등을 만들 때 사용가능함
- A <-> B 상관관계 표현하기에 편하다
    - 예시: A가 B에 몇 점을 줬다, A와 C가 친구이다

```python
df_movie = pd.read_csv("./movie_rating.csv")
df_movie.head()
```

```python
df_movie.pivot_table(
    values=["rating"],
    index=df_movie.critic,
    columns=df_movie.title,
    aggfunc="sum",
    fill_value=0,
)
```

```python
pd.crosstab(
    index=df_movie.critic,
    columns=df_movie.title,
    values=df_movie.rating,
    aggfunc="first",
).fillna(0)
```

```python
df_movie.groupby(["critic", "title"]).agg({"rating": "first"}).unstack()
```
