# matplotlib

### matplotlib overview

```python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X, Y)
# plt.show() -> 요즘은 없어도 보여줌
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

# figure 위에 하나하나씩 쌓아 그려준다
plt.plot(X_1, Y_1)  # 기본 -> line graph
plt.plot(X_2, Y_2)
plt.plot(range(100), range(100))
plt.show()
```

```python
import random

X_1= [random.randint(1, 100) for _ in range(100)]
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()  # x의 순서를 고려를 해야한다
```

### Figure & Subplot

- subplot의 순서를 grid로 작성

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

fig = plt.figure()
fig.set_size_inches(5, 2)
# 하나의 figure에 여러 subplot 그린다
ax_1 = fig.add_subplot(1, 2, 1)  # (row 수, column 수, 위치)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1)
ax_2.plot(X_2, Y_2)

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(5, 5)  # 싸이즈 설정
# plt.style.use("ggplot")    # 스타일적용

ax = []
colors = ["b", "g", "r", "c", "m", "y", "k"]
for i in range(1, 7):
    ax.append(fig.add_subplot(2, 3, i))  # 두개의 plot 생성
    X_1 = np.arange(50)
    Y_1 = np.random.rand(50)
    c = colors[np.random.randint(1, len(colors))]

    ax[i - 1].plot(X_1, Y_1, c=c)
```

```python
fig = plt.figure()  # figure 반환
fig.set_size_inches(10, 4)  # 크기지정
ax_1 = fig.add_subplot(1, 2, 1)  # 두개의 plot 생성
ax_2 = fig.add_subplot(1, 2, 2)  # 두개의 plot 생성

ax_1.plot(X_1, Y_1, c="b")  # 첫번째 plot
ax_2.plot(X_2, Y_2, c="g")  # 두번째 plot
plt.show()  # show & flush
```

### Set color
- http://matplotlib.org/2.0.2/api/colors_api.html
- `color` 속성을 사용
- float: 흑백, rgb color, predefined color 사용

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color="#000000")  # rgb color
plt.plot(X_2, Y_2, c="c")

plt.show()
```

### Set linestyle
- https://matplotlib.org/examples/lines_bars_and_markers/linestyles.html
- `ls` 또는 `linestype` 속성 사용
    

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", ls="dotted")

plt.show()
```

### set legend

- legend 함수로 범례를 표시함
- `loc` 위치등 속성 지정

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label="line_1")
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label="line_2")
plt.legend(shadow=True, fancybox=False, loc="upper right")
# shadow - 범례에 그림자 유무
# fancybox - 3D처럼 보일지 말지
# loc - 위치 선정

plt.title("$y = ax+b$")
plt.xlabel("$x_line$")
plt.ylabel("y_line")


plt.show()
```

### set grid & xylim

- Graph 보조선을 긋는 grid와 xy축 범위 한계를 지정

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label="line_1")
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label="line_2")
plt.legend(shadow=True, fancybox=True, loc="lower right")


plt.grid(True, lw=0.4, ls="--", c=".90")
plt.xlim(-1000, 2000)  # x의 최대, 최소
plt.ylim(-1000, 2000)  # y의 최대, 최소

plt.show()
```

#### 월별 통화량 데이터 총합 구해보기

```python
import pandas as pd

df = pd.read_csv("./phone_data.csv")
df.head()
```

```python
result = df.groupby(["month"])["duration"].sum()
result
```

```python
plt.plot(result.index, result, color="b", linestyle="dashed")

plt.title("Durations per month")
plt.show()
```

#### 일별 네트워크별 데이터의 총량을 다중 plot으로 작성하라

```python
def parsedate(date_info):
    import datetime

    return date_info[:9]
```

```python
df["day"] = df["date"].map(parsedate)
df.head()
```

```python
# pandas에서 duration에 대한 데이터를 network, day를 기준으로 뽑는다
result = df.groupby(["network", "day"])["duration"].sum().reset_index()
result.head()
```

```python
result["network"].unique().tolist()
```

```python
result[result["network"] == "data"].head()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)  # 싸이즈 설정
# plt.style.use("ggplot")    # 스타일적용

network_types = result["network"].unique().tolist()
ax = []
for i in range(1, 7):
    ax.append(fig.add_subplot(2, 3, i))  # 두개의 plot 생성
    network_name = network_types[i - 1]

    plt.title(network_name)

    X_1 = result[result["network"] == network_name]["day"]
    Y_1 = result[result["network"] == network_name]["duration"]
    ax[i - 1].get_xaxis().set_visible(False)
    ax[i - 1].plot(X_1, Y_1)
```

### Scatter

- `scatter` 함수 사용, `marker` : scatter 모양 지정

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
```

```python
# plt.scatter(x, y, color, 모양)
# line 그래프와 다르게 sorting할 필요 없다
plt.scatter(data_1[:, 0], data_1[:, 1], c="b", marker="x")
plt.scatter(data_2[:, 0], data_2[:, 1], c="r", marker="o")

plt.show()
```

```python
result = df.groupby(["network", "day"])["duration"].sum().reset_index()
networks = result["network"].unique().tolist()
colors = ["b", "g", "r", "c", "m", "y", "k"]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

for network_name, color in zip(networks, colors):
    if network_name == "data":
        continue
    X_1 = result[result["network"] == network_name]["day"]
    Y_1 = result[result["network"] == network_name]["duration"]

    # ax에 scatter로 데이터를 계속 쌓아준다
    ax.scatter(X_1, Y_1, c=color, label=network_name)
    ax.get_xaxis().set_visible(False)

ax.set_xlabel("day")
ax.set_ylabel("duration")

plt.legend(shadow=True, fancybox=True, loc="upper right")
plt.show()  # flush
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2

# s : 데이터의 크기를 지정, 데이터의 크기비교 가능
# s에도 sequence형 data 넣는다
# alpha : 투명도
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
len(x), len(y), len(area)  # 모두 같아야 하고, 모두 sequence형 데이터
```

### Bar chart

- `bar` 함수 사용
- 실전에서 잘 쓰지 않는다고 한다

```python
data = [[5.0, 25.0, 50.0, 20.0], [4.0, 23.0, 51.0, 17], [6.0, 22.0, 52.0, 19]]

X = np.arange(0, 8, 2)
```

```python
# X + 0.00 -> broadcasting

plt.bar(X + 0.00, data[0], color="b", width=0.50)
plt.bar(X + 0.50, data[1], color="g", width=0.50)
plt.bar(X + 1.0, data[2], color="r", width=0.50)
plt.xticks(X + 0.50, ("A", "B", "C", "D"))  # 글자 보여준다, sequence형 자료형 
plt.show()
```

```python
data = np.array([[5.0, 25.0, 50.0, 20.0], [4.0, 23.0, 51.0, 17], [6.0, 22.0, 52.0, 19]])

color_list = ["b", "g", "r"]
data_label = ["A", "B", "C"]
X = np.arange(data.shape[1])
X
```

```python
data = np.array([[5.0, 5.0, 5.0, 5.0], [4.0, 23.0, 51.0, 17], [6.0, 22.0, 52.0, 19]])

# bottom -> [5, 4, 6]에서 6 그릴 때 5+4=9에서부터 시작
for i in range(3):
    plt.bar(
        X,
        data[i],
        bottom=np.sum(data[:i], axis=0),
        color=color_list[i],
        label=data_label[i],
    )
plt.legend()
plt.show()
```

```python
A = [5.0, 30.0, 45.0, 22.0]
B = [5, 25, 50, 20]

X = range(4)

plt.bar(X, A, color="b")
plt.bar(X, B, color="r", bottom=60)
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)  # 여기서는 y축

plt.barh(X, women_pop, color="r")  # bar horizontal
plt.barh(X, -men_pop, color="b")
plt.show()
```
